```python
labels_f = open('labels_for_android.txt', 'a')
labels_f.write("<string-array name=\"sign_labels\">\n")
for label in labels:
    label = label.replace('-',' ')
    label = label.replace('  ', ' ')
    label = label[:len(label) - 3]
    labels_f.write(f"\t<item>{label.replace('-',' ')}</item>\n")
labels_f.write("</string-array>\n")
```
